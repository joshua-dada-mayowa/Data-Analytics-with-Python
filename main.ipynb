```python
import opendatasets as od
url='https://www.kaggle.com/datasets/usmanlovescode/nigeria-food-prices-dataset'

od.download(url)
```

```python
datafile='./nigeria-food-prices-dataset/wfp_food_prices_nga.csv'
```

```python
import pandas as pd

food_price_df=pd.read_csv(datafile)
```

```python
food_price_df.head(20)
```

```python
food_price_df.describe()
```

```python
food_price_df.rename(columns={'admin1': 'State','admin2':'City','commodity':'FoodType','category':'FoodCategory','pricetype':'PriceType','priceflag':'PriceFlag'}, inplace=True)
```

```python
food_price_df.dtypes
```

```python
food_price_df.columns
```


Selecting Specific Columns for Analysis

```python
food_price_df=food_price_df[['date', 'State', 'City', 'market',
       'FoodCategory', 'FoodType', 'unit', 'PriceFlag', 'PriceType','price']][1:]
```

Converting date coln(s) to datetime datatypes

```python
food_price_df['date']=pd.to_datetime(food_price_df['date'],yearfirst=True)

```

Converting numeric colns to numeric datatypes

```python
food_price_df['price']=pd.to_numeric(food_price_df['price']).round(2)
food_price_df.dtypes
```

Reseting Index

```python
food_price_df.reset_index(drop=True)
```

```python
food_price_df.to_csv('sample.csv')
```

```python
# Extract year and month for more granular analysis
food_price_df['year'] = food_price_df['date'].dt.year

# Group by year, FoodCategory, and FoodType to calculate the average price
average_prices = food_price_df.groupby(['year', 'FoodType'])['price'].mean().reset_index()
average_prices
```

```python
food_type= average_prices['FoodType'].unique()
```

```python
for food_type in food_type:
    food_data = average_prices[average_prices['FoodType'] == food_type]
    plt.plot(food_data['year'], food_data['price'], label=f"{food_type}")

plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('Average Food Prices Over Time by Food Category and Type')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.figure(figsize=(55,10))
plt.show()
```

```python
# Count the occurrences of each FoodType
food_type_counts = food_price_df['FoodType'].value_counts().reset_index()
food_type_counts.columns = ['FoodType', 'Count']

# Display the food type with the highest count
most_popular_food_type = food_type_counts.iloc[0]
food_type_counts, most_popular_food_type
```

# Analysis

### Average Price Trends Over Time By Food Category

```python
price_trends = food_price_df.groupby(['date', 'FoodCategory'])['price'].mean().unstack()
```

- Visualization

The plot shows the average price trends over time for different food categories. 

Each line represents a food category, illustrating how prices have fluctuated over the years.

```python
plt.figure(figsize=(35, 30))
price_trends.plot(ax=plt.gca())
plt.title('Average Price Trends Over Time by Food Category')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.legend(title='Food Category')
plt.grid(True)
plt.show()
```

### Average Price Per Unit By Food Category And Type

```python
price_per_unit = food_price_df.groupby(['FoodCategory', 'FoodType'])['price'].mean().reset_index()
price_per_unit.head()
```

### Average Price Trends Over Time For Top 10 Food Types

```python
top_10_food_types = food_price_df['FoodType'].value_counts().nlargest(10).index

top_10_data = food_price_df[food_price_df['FoodType'].isin(top_10_food_types)]

price_trends_top_10 = top_10_data.groupby(['date', 'FoodType'])['price'].mean().unstack()

price_trends_top_10
```

- Visualization

The plot displays the average price trends over time for the top 10 most popular food types. 

Each line represents a specific food type, showing how its price has changed over the years.

```python
plt.figure(figsize=(14, 7))
price_trends_top_10.plot(ax=plt.gca())
plt.title('Average Price Trends Over Time for Top 10 Food Types')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.legend(title='Food Type')
plt.grid(True)
plt.show()
```

```python
average_price_by_market = food_price_df.groupby('market')['price'].mean().reset_index()
average_price_by_market = average_price_by_market.sort_values(by='price', ascending=False)
```

- Top Markets by Average price

```python
top_markets = average_price_by_market[['market','price']].round(2)

top_markets
```

### Top Food Item and Category by Market

```python
def get_top_food_item_category(market_data):
    top_food_item = market_data.groupby(['FoodCategory', 'FoodType'])['price'].mean().idxmax()
    return pd.Series(top_food_item, index=['TopFoodCategory', 'TopFoodType'])
```

```python
top_food_item_category_by_market =food_price_df.groupby('market').apply(get_top_food_item_category).reset_index()

top_food_item_category_by_market = top_food_item_category_by_market[top_food_item_category_by_market['market'].isin(top_markets['market'])]

top_food_item_category_by_market
```
